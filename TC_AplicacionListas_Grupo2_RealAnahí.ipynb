## **1. Verificador de Expresiones con Paréntesis**
Dado un string que contiene una expresión matemática o de código, determinar si los paréntesis, corchetes y llaves están correctamente balanceados. Este es un problema clásico resuelto eficientemente con una pila.


```python
# Definición de la clase Nodo para la lista enlazada
class Nodo:
    def __init__(self, dato):
        self.dato = dato  # El dato que almacena el nodo
        self.siguiente = None  # Referencia al siguiente nodo en la lista

# Implementación de una Pila (Stack) usando una lista enlazada simple
class PilaEnlazada:
    """Pila implementada sobre lista enlazada simple."""

    def __init__(self):
        self.tope = None  # Puntero al elemento superior de la pila
        self._tamanio = 0  # Contador para el tamaño de la pila

    def push(self, dato):
        # Agrega un elemento a la parte superior de la pila
        nuevo = Nodo(dato)  # Crea un nuevo nodo con el dato
        nuevo.siguiente = self.tope  # El nuevo nodo apunta al antiguo tope
        self.tope = nuevo  # El nuevo nodo se convierte en el tope
        self._tamanio += 1  # Incrementa el tamaño de la pila

    def pop(self):
        # Elimina y devuelve el elemento superior de la pila
        if self.esta_vacia():
            raise IndexError("pop en pila vacía") # Error si la pila está vacía
        valor = self.tope.dato  # Guarda el dato del tope
        self.tope = self.tope.siguiente  # Mueve el tope al siguiente nodo
        self._tamanio -= 1  # Decrementa el tamaño de la pila
        return valor

    def peek(self):
        # Devuelve el elemento superior de la pila sin eliminarlo
        if self.esta_vacia():
            return None  # Retorna None si la pila está vacía
        return self.tope.dato

    def esta_vacia(self):
        # Verifica si la pila está vacía
        return self.tope is None

    def __len__(self):
        # Permite usar len() para obtener el tamaño de la pila
        return self._tamanio

# Función para verificar si los delimitadores (paréntesis, corchetes, llaves) en una expresión están balanceados
def verificar_expresion(expresion: str) -> bool:
    """
    Verifica si los delimitadores de una expresión están balanceados.
    Args:
    expresion: Cadena de texto a evaluar.
    Returns:
    True si está balanceada, False en caso contrario.
    """
    pila = PilaEnlazada()  # Inicializa una pila para guardar los delimitadores de apertura
    apertura = set('([{')  # Conjunto de delimitadores de apertura
    cierre = set(')]}')  # Conjunto de delimitadores de cierre
    pares = {')': '(', ']': '[', '}': '{'}  # Mapeo de delimitadores de cierre a apertura

    for char in expresion:
        if char in apertura:
            pila.push(char)  # Si es de apertura, lo agrega a la pila
        elif char in cierre:
            # Si es de cierre, verifica si la pila está vacía o si el tope no coincide
            if pila.esta_vacia() or pila.pop() != pares[char]:
                return False  # La expresión no está balanceada

    return pila.esta_vacia()  # La expresión está balanceada si la pila está vacía al final

# ── Pruebas ──────────────────────────────────────────────
if __name__ == '__main__':
    # Casos de prueba para la función verificar_expresion
    casos = [
        ("({[a + b] * (c - d)})", True),
        ("({[})", False),
        ("(((((", False),
        ("int x = (a + b);", True),
        ("", True),
    ]
    print('Ejercicio 1 — Verificador de Expresiones')
    print('-' * 45)
    # Itera sobre los casos de prueba y muestra los resultados
    for expr, esperado in casos:
        resultado = verificar_expresion(expr)
        estado = 'OK' if resultado == esperado else 'FALLO'
        print(f'[{estado}] "{expr}" → {resultado}')


```

Para esta extensión, crearemos una nueva función `verificar_expresion_con_comillas` que expandirá la lógica original para identificar y omitir cualquier carácter que se encuentre dentro de comillas simples o dobles.
```python
def verificar_expresion_con_comillas(expresion: str) -> bool:
    """
    Verifica si los delimitadores de una expresión están balanceados, ignorando el contenido entre comillas.
    Args:
    expresion: Cadena de texto a evaluar.
    Returns:
    True si está balanceada, False en caso contrario.
    """
    pila = PilaEnlazada()  # Inicializa una pila para guardar los delimitadores de apertura
    apertura = set('([{')  # Conjunto de caracteres que abren un delimitador
    cierre = set(')]}')  # Conjunto de caracteres que cierran un delimitador
    pares = {')': '(', ']': '[', '}': '{'}  # Mapea los delimitadores de cierre a sus correspondientes de apertura

    # Variables para saber si estamos dentro de comillas (simples o dobles)
    dentro_comilla_simple = False
    dentro_comilla_doble = False

    # Recorre cada carácter de la expresión
    for char in expresion:
        # Maneja las comillas simples
        if char == "'":
            # Si no estamos dentro de comillas dobles, alterna el estado de comilla simple
            if not dentro_comilla_doble:
                dentro_comilla_simple = not dentro_comilla_simple
            continue # Ignora el carácter de comilla, no es un delimitador a balancear

        # Maneja las comillas dobles
        if char == '"':
            # Si no estamos dentro de comillas simples, alterna el estado de comilla doble
            if not dentro_comilla_simple:
                dentro_comilla_doble = not dentro_comilla_doble
            continue # Ignora el carácter de comilla, no es un delimitador a balancear

        # Si estamos dentro de cualquier tipo de comilla, ignoramos los delimitadores
        if dentro_comilla_simple or dentro_comilla_doble:
            continue

        # Si el carácter es un delimitador de apertura
        if char in apertura:
            pila.push(char) # Lo agregamos a la pila
        # Si el carácter es un delimitador de cierre
        elif char in cierre:
            # Si la pila está vacía o el delimitador no coincide, la expresión no está balanceada
            if pila.esta_vacia() or pila.pop() != pares[char]:
                return False

    # Al final, la expresión está balanceada si la pila está vacía y no hemos dejado ninguna comilla abierta
    return pila.esta_vacia() and not dentro_comilla_simple and not dentro_comilla_doble

# ── Pruebas para la nueva función ──────────────────────────────
print('Ejercicio 2 — Verificador de Expresiones con Comillas')
print('-' * 60)

casos_comillas = [
    ("([a + b] * 'string con (paréntesis)' - c)", True), # Comillas simples
    ("([a + b] * \"string con [corchetes] y (paréntesis)\" - c)", True), # Comillas dobles
    ("({[a + b] * \"(c - d)}\")", False), # Delimitador de cierre fuera de comillas, pero el de apertura se ignoró
    ("(\"\"{\"\"}\"\")", True), # Comillas vacías y delimitadores balanceados
    ("([a + b] * 'string con (paréntesis)' - c])", False), # Delimitador desbalanceado
    ("hola' (mundo) '", True), # Comillas que contienen delimitadores y no afectan el balance
    ("('a)", False) # Comilla simple abierta al final
]

for expr, esperado in casos_comillas:
    resultado = verificar_expresion_con_comillas(expr)
    estado = 'OK' if resultado == esperado else 'FALLO'
    print(f'[{estado}] "{expr}" → {resultado}')
```

## **2.Sistema de Atención por Turnos (Cola de Prioridad)**
Simular el sistema de turnos de un banco o consultorio. Los clientes se registran con nombre y tipo (normal /
preferencial). Los clientes preferenciales se atienden antes que los normales, pero dentro de cada categoría se
respeta el orden de llegada (FIFO). Se implementa una cola separada para cada categoría

```python
from dataclasses import dataclass
from typing import Optional

class Nodo:
  def __init__(self, dato):
    self.dato = dato
    self.siguiente = None

class ColaEnlazada:
  """Cola (FIFO) implementada sobre lista enlazada simple."""

  def __init__(self):
    self.frente = None # primer nodo (para dequeue)
    self.final = None # último nodo (para enqueue)
    self._tamanio = 0

  def enqueue(self, dato):
    nuevo = Nodo(dato)
    # Si la cola está vacía, el nuevo nodo es el frente y el final
    if self.esta_vacia():
      self.frente = self.final = nuevo
    else:
      # Si no está vacía, el nuevo nodo se añade al final
      self.final.siguiente = nuevo
      self.final = nuevo
    self._tamanio += 1

  def dequeue(self):
    # Si la cola está vacía, lanza un error
    if self.esta_vacia():
      raise IndexError("dequeue en cola vacía")

    valor = self.frente.dato
    self.frente = self.frente.siguiente
    # Si el frente se vuelve None, significa que la cola ahora está vacía
    if self.frente is None:
      self.final = None
    self._tamanio -= 1
    return valor

  def front(self):
    # Retorna el dato del frente si la cola no está vacía, de lo contrario None
    return self.frente.dato if not self.esta_vacia() else None

  def esta_vacia(self):
    return self.frente is None

  def __len__(self):
    return self._tamanio

  def __str__(self):
    # Representación en cadena de la cola, mostrando sus elementos
    elementos = []
    actual = self.frente
    while actual:
      elementos.append(str(actual.dato))
      actual = actual.siguiente
    return "ColaEnlazada(" + ", ".join(elementos) + ")"

@dataclass
class Cliente:
  nombre: str
  tipo: str # 'preferencial' | 'normal'
  turno: int

  # Añadir método __str__ a la clase Cliente
  def __str__(self):
    return f"{self.nombre} (Turno: {self.turno}, Tipo: {self.tipo})"

class SistemaAtencion:
  """Sistema de turnos con dos colas: preferencial y normal."""
  def __init__(self):
    self.cola_preferencial = ColaEnlazada()
    self.cola_normal = ColaEnlazada()
    self._contador_turno = 1 # Inicializa el contador de turnos

  def registrar(self, nombre: str, tipo: str) -> int:
    """Agrega un cliente y retorna su número de turno."""
    turno = self._contador_turno
    cliente = Cliente(nombre=nombre, tipo=tipo, turno=turno)

    # Encola el cliente según su tipo
    if tipo == 'preferencial':
      self.cola_preferencial.enqueue(cliente)
    else:
      self.cola_normal.enqueue(cliente)
    self._contador_turno += 1 # Incrementa el contador para el siguiente cliente
    return turno

  def atender(self) -> Optional[Cliente]:
    """Atiende al próximo cliente según prioridad."""
    # Prioriza la cola preferencial
    if not self.cola_preferencial.esta_vacia():
      return self.cola_preferencial.dequeue()

    # Si no hay clientes preferenciales, atiende de la cola normal
    if not self.cola_normal.esta_vacia():
      return self.cola_normal.dequeue()

    # Si ambas colas están vacías, no hay clientes para atender
    return None

  def pendientes(self) -> int:
    # Retorna el número total de clientes pendientes
    return len(self.cola_preferencial) + len(self.cola_normal)

# ── Demo ─────────────────────────────────────────────────
if __name__ == '__main__':
  sistema = SistemaAtencion()

  llegadas = [
      ('Ana García', 'normal'),
      ('Luis Pérez', 'preferencial'),
      ('María Torres', 'normal'),
      ('Carlos Ruiz', 'preferencial'),
      ('Sofia Mendez', 'normal'),
      ('Elena Vargas', 'preferencial'),
  ]
  print('=== Registro de Clientes ===')
  for nombre, tipo in llegadas:
    turno = sistema.registrar(nombre, tipo)
    print(f' Turno {turno:02d} [{tipo:13s}] {nombre}')

  print()
  print('=== Orden de Atención ===')
  orden = 1
  while sistema.pendientes() > 0:
    cliente = sistema.atender()
    # Asegura que el cliente no sea None antes de imprimir
    if cliente:
      print(f' {orden}. {cliente.nombre} (turno {cliente.turno},{cliente.tipo})')
      orden += 1
```

```python
# Actividad de Extensión 2.5: Implementación de Extensiones al Sistema de Atención

# Las clases Nodo y ColaEnlazada (con su método __str__) están definidas en celdas anteriores y son reutilizadas aquí.

from dataclasses import dataclass
from typing import Optional

# Modificación de la clase Cliente para incluir el tipo 'urgente' (Actividad 4)
@dataclass
class Cliente:
  nombre: str
  tipo: str # 'urgente' | 'preferencial' | 'normal'
  turno: int

  def __str__(self):
    return f"{self.nombre} (Turno: {self.turno}, Tipo: {self.tipo})"

class SistemaAtencionExtendida: # Renombrado para evitar conflicto con la clase original
  """Sistema de turnos con tres colas: urgente, preferencial y normal.
  Incluye cálculo de tiempo de espera."""
  def __init__(self):
    self.cola_urgente = ColaEnlazada() # Actividad 4: Nueva cola 'urgente'
    self.cola_preferencial = ColaEnlazada()
    self.cola_normal = ColaEnlazada()
    self._contador_turno = 1
    self._tiempo_simulado_actual = 0 # Para el cálculo del tiempo de espera
    self._total_tiempo_espera = 0
    self._clientes_atendidos_count = 0

  def registrar(self, nombre: str, tipo: str) -> int:
    turno = self._contador_turno
    cliente = Cliente(nombre=nombre, tipo=tipo, turno=turno)

    if tipo == 'urgente':
      self.cola_urgente.enqueue(cliente)
    elif tipo == 'preferencial':
      self.cola_preferencial.enqueue(cliente)
    else:
      self.cola_normal.enqueue(cliente)
    self._contador_turno += 1
    return turno

  def atender(self, service_duration: int = 5) -> Optional[Cliente]: # Actividad 6: Preparación para cálculo de tiempo
    cliente_atendido = None

    # Actividad 4: Prioridad a la cola urgente
    if not self.cola_urgente.esta_vacia():
      cliente_atendido = self.cola_urgente.dequeue()
    elif not self.cola_preferencial.esta_vacia():
      cliente_atendido = self.cola_preferencial.dequeue()
    elif not self.cola_normal.esta_vacia():
      cliente_atendido = self.cola_normal.dequeue()

    if cliente_atendido:
      # Actividad 6: Cálculo del tiempo de espera
      arrival_time_unit = cliente_atendido.turno # Asumimos que el turno es una unidad de tiempo de llegada

      # El tiempo en que este cliente puede empezar a ser atendido es el máximo entre
      # el tiempo actual simulado (cuando el sistema está libre) y el tiempo de llegada del cliente
      service_start_time = max(self._tiempo_simulado_actual, arrival_time_unit)
      waiting_time = service_start_time - arrival_time_unit

      self._total_tiempo_espera += waiting_time
      self._clientes_atendidos_count += 1
      self._tiempo_simulado_actual = service_start_time + service_duration # Actualiza el tiempo simulado

    return cliente_atendido

  def pendientes(self) -> int:
    return len(self.cola_urgente) + len(self.cola_preferencial) + len(self.cola_normal)

  def get_average_waiting_time(self) -> float: # Actividad 6: Método para obtener el promedio de espera
    if self._clientes_atendidos_count == 0:
      return 0.0
    return self._total_tiempo_espera / self._clientes_atendidos_count

# Actividad 5: Demostración del método __str__ de ColaEnlazada (ya implementado en la clase original)
# La clase ColaEnlazada ya contiene un método __str__ para visualizar sus elementos.
# A continuación, se mostrará su uso al imprimir las colas.

if __name__ == '__main__':
  print("--- Demostración de Extensiones del Sistema de Atención ---")
  sistema_ext = SistemaAtencionExtendida()

  llegadas_ext = [
      ('Ana García', 'normal'),
      ('Luis Pérez', 'preferencial'),
      ('Pedro Sanchez', 'urgente'), # Cliente urgente
      ('María Torres', 'normal'),
      ('Carlos Ruiz', 'preferencial'),
      ('Sofia Mendez', 'normal'),
      ('Elena Vargas', 'preferencial'),
      ('Diana Rojas', 'urgente'), # Otro cliente urgente
      ('Juanito Alcachofa', 'normal')
  ]
  print('\n=== Registro de Clientes (Extendido) ===')
  for nombre, tipo in llegadas_ext:
    turno = sistema_ext.registrar(nombre, tipo)
    print(f' Turno {turno:02d} [{tipo:13s}] {nombre}')

  print('\n=== Visualización de Colas (Actividad 5) ===')
  # Uso del método __str__ de ColaEnlazada
  print(f'Cola Urgente: {sistema_ext.cola_urgente}')
  print(f'Cola Preferencial: {sistema_ext.cola_preferencial}')
  print(f'Cola Normal: {sistema_ext.cola_normal}')

  print('\n=== Orden de Atención (Extendido) ===')
  orden = 1
  service_duration_per_client = 5 # minutos por cliente (Actividad 6)
  while sistema_ext.pendientes() > 0:
    cliente_atendido = sistema_ext.atender(service_duration_per_client)
    if cliente_atendido:
      print(f' {orden}. {cliente_atendido.nombre} (turno {cliente_atendido.turno},{cliente_atendido.tipo})')
      orden += 1

  print(f'\n=== Tiempo Promedio de Espera (Actividad 6) ===')
  print(f'Tiempo promedio de espera: {sistema_ext.get_average_waiting_time():.2f} minutos')
```

## **3.Evaluador de Expresiones Postfijas (RPN)**

Implementar un evaluador de expresiones en notación postfija (Reverse Polish Notation). El proceso tiene dos fases: primero se tokeniza la expresión usando una cola (para respetar el orden izquierda-derecha), y luego se evalúa cada token usando una pila. Combina ambas estructuras en un mismo flujo.


```python
class Nodo:
    def __init__(self, dato):
        self.dato = dato
        self.siguiente = None

class PilaEnlazada:
    """Pila (LIFO) implementada sobre lista enlazada simple para el evaluador RPN."""
    def __init__(self):
        self.tope = None # Puntero al elemento superior de la pila

    def push(self, dato):
        """Agrega un elemento a la parte superior de la pila."""
        nuevo = Nodo(dato)
        nuevo.siguiente = self.tope
        self.tope = nuevo

    def pop(self):
        """Elimina y devuelve el elemento superior de la pila."""
        if self.esta_vacia():
            raise IndexError("pop en pila vacía")
        valor = self.tope.dato
        self.tope = self.tope.siguiente
        return valor

    def esta_vacia(self):
        """Verifica si la pila está vacía."""
        return self.tope is None

class ColaEnlazada:
    """Cola (FIFO) implementada sobre lista enlazada simple para tokenizar la expresión."""
    def __init__(self):
        self.frente = None # Primer nodo (para dequeue)
        self.final = None  # Último nodo (para enqueue)

    def enqueue(self, dato):
        """Agrega un elemento al final de la cola."""
        nuevo = Nodo(dato)
        if self.frente is None:
            self.frente = self.final = nuevo
        else:
            self.final.siguiente = nuevo
            self.final = nuevo

    def dequeue(self):
        """Elimina y devuelve el elemento del frente de la cola."""
        if self.esta_vacia():
            raise IndexError("dequeue en cola vacía")
        valor = self.frente.dato
        self.frente = self.frente.siguiente
        if self.frente is None:
            self.final = None
        return valor

    def esta_vacia(self):
        """Verifica si la cola está vacía."""
        return self.frente is None

OPERADORES = {
    '+': lambda a, b: a + b,
    '-': lambda a, b: a - b,
    '*': lambda a, b: a * b,
    '/': lambda a, b: a / b,
    '^': lambda a, b: a ** b,
}

def evaluar_postfija(expresion: str) -> float:
    """
    Evalúa una expresión en notación postfija (Reverse Polish Notation).

    El proceso primero tokeniza la expresión usando una cola para respetar el orden
    de izquierda a derecha, y luego evalúa cada token usando una pila.

    Args:
        expresion: Cadena de texto que contiene la expresión postfija,
                   con los tokens (números y operadores) separados por espacios.
                   Ejemplo: '3 4 + 2 *'

    Returns:
        El resultado numérico de la evaluación de la expresión.

    Raises:
        ValueError: Si la expresión es inválida (operandos insuficientes, token desconocido,
                    o expresión vacía).
    """
    cola_tokens = ColaEnlazada()
    # Tokeniza la expresión y encola cada token
    for token in expresion.split():
        cola_tokens.enqueue(token)

    pila_operandos = PilaEnlazada()
    # Procesa los tokens de la cola uno por uno
    while not cola_tokens.esta_vacia():
        token = cola_tokens.dequeue()
        if token in OPERADORES:
            # Si el token es un operador, desapila dos operandos
            if pila_operandos.esta_vacia():
                raise ValueError(f'Operandos insuficientes para {token!r}')
            b = pila_operandos.pop()
            if pila_operandos.esta_vacia():
                raise ValueError(f'Operandos insuficientes para {token!r}')
            a = pila_operandos.pop()
            # Realiza la operación y apila el resultado
            resultado = OPERADORES[token](a, b)
            pila_operandos.push(resultado)
        else:
            # Si el token es un número, lo convierte a float y lo apila
            try:
                pila_operandos.push(float(token))
            except ValueError:
                raise ValueError(f'Token desconocido: {token!r}')

    # Al final, la pila debe contener solo el resultado final
    if pila_operandos.esta_vacia():
        raise ValueError('Expresión vacía')
    return pila_operandos.pop()

# ── Pruebas ──────────────────────────────────────────────
if __name__ == '__main__':
    casos = [
        ('3 4 + 2 *', 14.0, '(3+4)*2'),
        ('5 1 2 + 4 * +', 17.0, '5+((1+2)*4)'),
        ('2 3 ^ 1 +', 9.0, '2^3+1'),
        ('10 2 /', 5.0, '10/2'),
        ('15 7 1 1 + - / 3 * 2 1 1 + + -', 5.0, '((15/(7-(1+1)))*3-(2+(1+1)))'),
    ]

    print('Ejercicio 3 — Evaluador de Expresiones Postfijas (RPN)')
    print('-' * 60)
    for postfija, esperado, infija in casos:
        resultado = evaluar_postfija(postfija)
        estado = 'OK' if abs(resultado - esperado) < 1e-9 else 'FALLO'
        print(f'[{estado}] {infija}')
        print(f' Postfija: {postfija}')
        print(f' Resultado: {resultado} (esperado: {esperado})')
        print()
```

```python
import math

# ==========================================
# EXTENSIONES (Actividades 7, 8 y 9)
# ==========================================

# ── Actividad 9: Funciones Unarias ──
FUNCIONES = {
    'sqrt': math.sqrt,
    'abs': abs
}

def evaluar_postfija_extendida(expresion: str) -> float:
    """Versión extendida que soporta funciones unarias como sqrt y abs."""
    cola_tokens = ColaEnlazada()
    for token in expresion.split():
        cola_tokens.enqueue(token)

    pila_operandos = PilaEnlazada()
    while not cola_tokens.esta_vacia():
        token = cola_tokens.dequeue()

        if token in OPERADORES:
            b = pila_operandos.pop()
            a = pila_operandos.pop()
            pila_operandos.push(OPERADORES[token](a, b))
        elif token in FUNCIONES:
            a = pila_operandos.pop() # Las funciones unarias solo sacan 1 operando
            pila_operandos.push(FUNCIONES[token](a))
        else:
            try:
                pila_operandos.push(float(token))
            except ValueError:
                raise ValueError(f'Token desconocido: {token!r}')

    return pila_operandos.pop()


# ── Actividad 7: Algoritmo Shunting-Yard ──
PRECEDENCIA = {'+': 1, '-': 1, '*': 2, '/': 2, '^': 3}

def infija_a_postfija(expresion: str) -> str:
    """
    Convierte una expresión infija (separada por espacios) a postfija.
    Ejemplo: '3 + 4 * 2' -> '3 4 2 * +'
    """
    pila_operadores = PilaEnlazada()
    salida = ColaEnlazada()

    for token in expresion.split():
        # Si es un número (incluyendo negativos/decimales) o una función
        if token.replace('.', '', 1).replace('-', '', 1).isdigit() or token in FUNCIONES:
            salida.enqueue(token)
        elif token == '(':
            pila_operadores.push(token)
        elif token == ')':
            while not pila_operadores.esta_vacia() and pila_operadores.tope.dato != '(':
                salida.enqueue(pila_operadores.pop())
            if not pila_operadores.esta_vacia() and pila_operadores.tope.dato == '(':
                pila_operadores.pop() # Descartar el '('
        elif token in PRECEDENCIA:
            while (not pila_operadores.esta_vacia() and
                   pila_operadores.tope.dato != '(' and
                   PRECEDENCIA.get(pila_operadores.tope.dato, 0) >= PRECEDENCIA[token]):
                salida.enqueue(pila_operadores.pop())
            pila_operadores.push(token)

    # Vaciar los operadores restantes en la pila
    while not pila_operadores.esta_vacia():
        salida.enqueue(pila_operadores.pop())

    # Reconstruir el string final
    resultado = []
    while not salida.esta_vacia():
        resultado.append(salida.dequeue())

    return " ".join(resultado)


# ── Pruebas de las Extensiones ──────────────────────────────────────────
if __name__ == '__main__':
    # Mantengo tus casos originales ejecutándose primero (omitidos aquí por brevedad)

    print('\n' + '=' * 60)
    print('PRUEBAS DE EXTENSIONES (Actividades 7, 8 y 9)')
    print('=' * 60)

    # Actividad 8: Decimales y Negativos
    print("\n--- Probando Decimales y Negativos ---")
    postfija_dec = "-3.5 2.5 +"
    print(f"Expresión postfija: {postfija_dec}")
    print(f"Resultado: {evaluar_postfija(postfija_dec)} (Esperado: -1.0)")

    # Actividad 9: Funciones Unarias (sqrt, abs)
    print("\n--- Probando Funciones (sqrt, abs) ---")
    postfija_func = "16 sqrt -5 abs +"
    print(f"Expresión postfija: {postfija_func}")
    print(f"Resultado: {evaluar_postfija_extendida(postfija_func)} (Esperado: 9.0)")

    # Actividad 7: Shunting-Yard (Infija a Postfija)
    print("\n--- Probando Shunting-Yard (Infija a Postfija) ---")
    casos_shunting = [
        ("3 + 4 * 2 / ( 1 - 5 ) ^ 2", "3 4 2 * 1 5 - 2 ^ / +"),
        ("abs ( -10 ) + sqrt ( 25 )", "-10 abs 25 sqrt +")
    ]

    for infija, postfija_esperada in casos_shunting:
        obtenida = infija_a_postfija(infija)
        print(f"Infija original : {infija}")
        print(f"Postfija obtenida: {obtenida}")
        if obtenida == postfija_esperada: # Evaluamos usando la función extendida
            print(f"Evaluación final : {evaluar_postfija_extendida(obtenida)}")
        print()
```
